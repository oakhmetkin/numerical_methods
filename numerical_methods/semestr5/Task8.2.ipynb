## Исходная функция f

$ y' = f(x) $

```python
def f(x: float, y: np.array):
    y1, y2, y3, y4 = y
    
    assert B == 3/2, '1/B = 2/3 for the first value' 
    
    return np.array([
        2*x * np.cbrt(y2)**2 * y4, # 2*x * y2**(1/B) * y4
        2*B*x * np.exp(B/C * (y3 - A)) * y4,
        2*C*x * y4,
        -2*x * np.log(max(1e-1, y1)), # if y1 <= 0
    ])
```

## Реализация метода Рунге-Кутта (2-этапный)

```python
def runge_kutta2(f, x0: float, y0: np.array, h: float, x_max: float, N_max=1000):
    X, Y = [x0], [y0]
    errors = [(0, 0, 0, 0)]
    
    x, y = x0, y0
    iterations = 0
    
    while x < x_max - 1e-4 >= 1e-6 and iterations < N_max:
        y = y + h*f(x + h/2, y + h/2 * f(x, y))
        
        h_half = h / 2
        y_half = y + h_half*f(x + h_half/2, y + h_half/2 * f(x, y))
        
        x = x + h
        
        X.append(x)
        Y.append(y)
        errors.append(abs(y - y_half))
        
        iterations += 1
    
    return np.array(X), np.array(Y), np.array(errors), iterations
```

## Реализация метода Рунге-Кутта (3-этапный) - метод оппонент

```python
def runge_kutta3(f, x0: float, y0: np.array, h: float, x_max: float, N_max=1000):
    X, Y = [x0], [y0]
    errors = [(0, 0, 0, 0)]
    
    x, y = x0, y0
    iterations = 0
    
    while x < x_max - 1e-4 >= 1e-6 and iterations < N_max:
        k1 = f(x, y)
        k2 = f(x + 1/2*h, y + 1/2*h*k1)
        k3 = f(x + h, y - h*k1 + 2*h*k2)
        y = y + h*(1/6*k1 + 4/6*k2 + 1/6*k3)
        
        h_half = h / 2
        k1 = f(x, y)
        k2 = f(x + 1/2*h_half, y + 1/2*h_half*k1)
        k3 = f(x + h_half, y - h_half*k1 + 2*h_half*k2)
        y_half = y + h_half*(1/6*k1 + 4/6*k2 + 1/6*k3)
        
        x = x + h
        
        X.append(x)
        Y.append(y)
        errors.append(abs(y - y_half))
        
        
        iterations += 1
    
    return np.array(X), np.array(Y), np.array(errors), iterations
```

## Точное решение системы дифф. уравнений

```python
def get_Y_exact(x):
    return np.array([
        np.exp(np.sin(x**2)),
        np.exp(B * np.sin(x**2)),
        C * np.sin(x**2) + A,
        np.cos(x**2),
    ])
```

## Вычисление решений и сравнение

$ h = 0.1 $

```python
x0, x_max = 0, 5
y0 = [1, 1, A, 1]
h = 0.1

# exact solution
X_exact = np.linspace(x0, x_max, 200)
Y_exact = get_Y_exact(X_exact).transpose(1, 0)

# Runge-Kutta 2-step
X2, Y2, errors2, iterations2 = runge_kutta2(f, x0, y0, h, x_max)
print(f'Метод Рунге-Кутты 2-шаговый: {iterations2} итераций')

# Runge-Kutta 3-step
X3, Y3, errors3, iterations3 = runge_kutta3(f, x0, y0, h, x_max)
print(f'Метод Рунге-Кутты 3-шаговый: {iterations3} итераций')
```

```python
fig, axs = plt.subplots(2, 2, figsize=(14, 8))

sol_texts = [
    r'e^{\sin{x^2}}',
    r'e^B{\sin{x^2}}',
    r'C \sin{x^2} + A',
    r'\cos{x^2}',
]

for i, ax in enumerate(axs.flatten()):
    ax.plot(X_exact, Y_exact[:, i], label='exact solution')
    ax.plot(X2, Y2[:, i], label='Runge-Kutta 2-step')
    ax.plot(X3, Y3[:, i], label='Runge-Kutta 3-step')
    ax.set_title(f'$ y_{i+1} = {sol_texts[i]} $')
    ax.legend()
```

## Автоматический выбор шага

```python
def runge_kutta2_auto_h(f, x0: float, y0: np.array, h_max: float, x_max: float, N_max=1000, eps=0.1):
    X, Y = [x0], [y0]
    errors = [(0, 0, 0, 0)]
    H = []
    
    x, y = x0, y0
    h = h_max
    iterations = 0
    
    eps1 = eps * 2
    eps2 = eps
    eps3 = eps / 4
    
    while x < x_max - 1e-4 and iterations < N_max:
        y_half = y + (h/2)*f(x + (h/2)/2, y + (h/2)/2 * f(x, y))
        y_full = y + h*f(x + h/2, y + h/2 * f(x, y))
        loc_err = max(abs(y_full - y_half))
        
        if h < 1e-6:
            h = 1e-6
            continue
        elif eps1 < loc_err:
            h /= 2
            continue
        
        y_half = y + (h/2)*f(x + (h/2)/2, y + (h/2)/2 * f(x, y))
        y = y + h*f(x + h/2, y + h/2 * f(x, y))
        x = x + h
        X.append(x)
        Y.append(y)
        H.append(h)
        errors.append(abs(y_half - y))
        
        if eps2 < loc_err < eps1:
            h /= 2
        elif eps3 < loc_err < eps2:
            pass
        elif loc_err < eps3:
            h = min(2*h, h_max)
        
        iterations += 1
    
    return np.array(X), np.array(Y), np.array(H), np.array(errors), iterations
```

```python
def rk3_get_for(x, y, h):
    k1 = f(x, y)
    k2 = f(x + 1/2*h, y + 1/2*h*k1)
    k3 = f(x + h, y - h*k1 + 2*h*k2)
    y = y + h*(1/6*k1 + 4/6*k2 + 1/6*k3)
    return y


def runge_kutta3_auto_h(f, x0: float, y0: np.array, h_max: float, x_max: float, N_max=1000, eps=0.1):
    X, Y = [x0], [y0]
    errors = [(0, 0, 0, 0)]
    H = []
    
    x, y = x0, y0
    h = h_max
    iterations = 0
    
    eps1 = eps * 2
    eps2 = eps
    eps3 = eps / 4
    
    while x < x_max - 1e-4 >= 1e-6 and iterations < N_max:
        y_full = rk3_get_for(x, y, h)
        y_half = rk3_get_for(x, y, h / 2)
        loc_err = max(abs(y_full - y_half))
        
        if h < 1e-6:
            h = 1e-6
            continue
        elif eps1 < loc_err:
            h /= 2
            continue
        
        y_half = rk3_get_for(x, y, h / 2)
        y = rk3_get_for(x, y, h)
        x = x + h
        X.append(x)
        Y.append(y)
        H.append(h)
        errors.append(abs(y_half - y))
        
        if eps2 < loc_err < eps1:
            h /= 2
        elif eps3 < loc_err < eps2:
            pass
        elif loc_err < eps3:
            h = min(2*h, h_max)
        
        iterations += 1
    
    return np.array(X), np.array(Y), np.array(H), np.array(errors), iterations
```

```python
x0, x_max = 0, 5
y0 = [1, 1, A, 1]
h = 0.5

# exact solution
X_exact = np.linspace(x0, x_max, 200)
Y_exact = get_Y_exact(X_exact).transpose(1, 0)

# Runge-Kutta 2-step auto h
X2, Y2, H2, errors2, iterations2 = runge_kutta2_auto_h(f, x0, y0, h, x_max, N_max=1000, eps=0.05)
print(f'Метод Рунге-Кутты 2-шаговый с автомат. выбором шага: {iterations2} итераций')

# Runge-Kutta 3-step auto h
X3, Y3, H3, errors3, iterations3 = runge_kutta3_auto_h(f, x0, y0, h, x_max, N_max=1000, eps=0.05)
print(f'Метод Рунге-Кутты 3-шаговый с автомат. выбором шага: {iterations3} итераций')
```
